```python
from bs4 import BeautifulSoup as bsp
import urllib
import http.cookiejar
import re
import time
```

## 空气质量数据
### 爬虫参考：https://blog.csdn.net/sinat_27339001/article/details/72956360
### 数据来源：天气后报 http://www.tianqihoubao.com/aqi/
* 爬取2014年1月至2018年11月各省会城市、直辖市空气质量情况（无港澳台数据）


```python
# 做好cookie管理工作
cookie=http.cookiejar.CookieJar() # 创建空CookieJar
cj=urllib.request.HTTPCookieProcessor(cookie) # 构造cookie
opener = urllib.request.build_opener(cj) # 根据cookie构造opener
# 伪造header
opener.addheaders = [('User-agent', 'Mozilla/4.0 (compatible; MSIE 7.0; Windows NT 5.1)')]
# 载入header
urllib.request.install_opener(opener)

```

```python
city_names=['beijing', 'tianjin', 'shijiazhuang', 'taiyuan', 'huhehaote', 'shenyang', 'changchun',
           'haerbin', 'shanghai', 'nanjing', 'hangzhou', 'hefei', 'fujianfuzhou', 'nanchang', 'jinan',
           'zhengzhou', 'wuhan', 'changsha', 'guangzhou', 'nanning', 'haikou', 'chongqing', 'chengdu',
           'guiyang', 'kunming', 'lasa', 'xian', 'lanzhou', 'xining', 'yinchuan', 'wulumuqi']
# seems no data for Hongkong、Mocao、Taibei
# 'taibei', 'xianggang', 'aomen']
#city_names = ['beijing']

years = ['2014', '2015', '2016', '2017', '2018']
#years = ['2018']
months = ['01', '02', '03', '04', '05', '06', '07', '08', '09', '10', '11', '12']
#months = ['01']
baseurl = 'http://www.tianqihoubao.com/aqi/'
# url should be: http://www.tianqihoubao.com/aqi/beijing-201405.html
```

```python
for city_name in city_names:
    time.sleep(2)
    file = open('./data/aqi/' + city_name + '.csv', 'w')
    file.write('日期,质量等级,AQI指数,当天AQI排名,PM2.5,PM10,So2,No2,Co,O3\n')
    
    for year in years:
        for month in months:
            # skip 201812
            if (year == '2018' and month == '12'):
                continue
                
            url = baseurl+city_name+'-'+year+month+'.html'
            
            myrequest = urllib.request.Request(url)
            #connection failed, sleep for 10s
            while True:
                try:
                    html = bsp(bytes.decode(urllib.request.urlopen(myrequest).read(),'gbk'))
                    break
                except:
                    print('connection failed')
                    time.sleep(10)
            
            # re: regular expression, sub for replace invisible character with empty
            u1=[re.sub('\s','',td.text) for td in html.table.find_all('td')]

            # skip head when writeing
            # only need one head for each city, already written
            # 10 for len(head)
            for j,item in enumerate(u1[10:]):
                if j%10==9:
                    file.write(item+'\n')
                else:
                    file.write(item+',')
            if (month == '06' or month == '12'):
                print('over: '+ url)
    file.close()
print('all over')
```
